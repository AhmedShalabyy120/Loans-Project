# Data Cleaning and Transformation

```python
def bin_columns(df, column_name, new_column_name):
    """
    Bins a specified column in the DataFrame into predefined categories.

    Parameters:
        df (pd.DataFrame): The DataFrame containing the column to bin.
        column_name (str): The name of the column to bin.
        new_column_name (str): The name of the new column for binned categories.

    Returns:
        pd.DataFrame: The updated DataFrame with the new binned column.
    """
    bins = [0, 20000, 50000, 100000, 200000, float('inf')]  # Define bins
    labels = ['0-20k', '20k-50k', '50k-100k', '100k-200k', '200k+']  # Define labels

    df[new_column_name] = pd.cut(df[column_name], bins=bins, labels=labels)  # Binning
    return df
```

```python
# Call the function with your existing DataFrame
df = bin_columns(df, column_name='annual_inc', new_column_name='income_category')

```

```python
df.income_category.value_counts()
```

```python
def bin_loan_amount(df, column_name, new_column_name):
    """
    Bins a specified loan amount column in the DataFrame into predefined categories.

    Parameters:
        df (pd.DataFrame): The DataFrame containing the loan amount column to bin.
        column_name (str): The name of the column to bin.
        new_column_name (str): The name of the new column for binned loan amounts.

    Returns:
        pd.DataFrame: The updated DataFrame with the new binned column.
    """
    loan_bins = [1000, 10000, 20000, 30000, 40000]  # Define loan ranges
    loan_labels = ['1k-10k', '10k-20k', '20k-30k', '30k-40k']  # Define labels

    df[new_column_name] = pd.cut(df[column_name], bins=loan_bins, labels=loan_labels)  # Binning
    return df

```

```python
# Call the function with your existing DataFrame
df = bin_loan_amount(df, column_name='loan_amount', new_column_name='loan_amount_category')
```

```python
def bin_grades(df, column_name, new_column_name):
    """
    Bins a specified grade column in the DataFrame into predefined categories.

    Parameters:
        df (pd.DataFrame): The DataFrame containing the grade column to bin.
        column_name (str): The name of the column to bin.
        new_column_name (str): The name of the new column for binned grades.

    Returns:
        pd.DataFrame: The updated DataFrame with the new binned column.
    """
    bins = [0, 5, 10, 15, 20, 25, 30, 35]  # Define grade ranges
    labels = ['Excellent', 'Good', 'Average', 'Below Average', 'Poor', 'Very Poor', 'Critical']  # Define labels

    df[new_column_name] = pd.cut(df[column_name], bins=bins, labels=labels, right=True)  # Binning
    return df
```

```python
# Call the function with your existing DataFrame
df = bin_grades(df, column_name='grade', new_column_name='grade_label')
```

# Adding Monthly installments

```python
# Define a function to calculate the monthly installment
def calculate_installment(row):
    P = row['loan_amount']  # Principal loan amount
    r = row['int_rate'] / 12  # Monthly interest rate
    n = row['term'] * 12  # Number of payments (convert term in years to months)

    if r == 0:  # Handle case with zero interest rate
        return P / n
    else:
        return (P * r * (1 + r)**n) / ((1 + r)**n - 1)

# Add a new column 'installment' to the DataFrame
df['installment'] = df.apply(calculate_installment, axis=1)
```

```python
def enrich_state_data(dataframe):
    """
    Enriches the DataFrame with full state names using FreeTestAPI
    and additional information from Statistics Canada.

    Args:
        dataframe (pd.DataFrame): DataFrame containing a column 'state' with state abbreviations.

    Returns:
        pd.DataFrame: DataFrame with an additional 'state_name' column.
    """
    import requests

    # 1. Fetch state names from FreeTestAPI
    api_endpoint = "https://freetestapi.com/api/v1/us-states"
    api_response = requests.get(api_endpoint)
    if api_response.status_code != 200:
        raise Exception("Failed to fetch state data from FreeTestAPI.")
    state_data_api = api_response.json()
    state_mapping_api = {entry['abbreviation']: entry['name'] for entry in state_data_api}

    # 2. Fetch state data from Statistics Canada
    statcan_endpoint = "https://www23.statcan.gc.ca/imdb/p3VD.pl?Function=getVD&TVD=53971"
    statcan_table = pd.read_html(statcan_endpoint)[0]  # Extract the first table
    statcan_table.columns = ['Code', 'Province/Territory', 'Abbreviation', 'Alpha Code']  # Set column names
    statcan_mapping = dict(zip(statcan_table['Alpha Code'], statcan_table['Province/Territory']))

    # 3. Convert 'state' column values to uppercase
    dataframe['state'] = dataframe['state'].str.upper()

    # 4. Map state abbreviations to full names from the API
    dataframe['state_full_name'] = dataframe['state'].map(state_mapping_api)

    # 5. Enhance state names with Statistics Canada data if available
    dataframe['state_full_name'] = dataframe['state'].map(statcan_mapping).fillna(dataframe['state_full_name'])

    # Save the result to a CSV file
    dataframe[['state', 'state_full_name']].to_csv('state_data_enriched.csv', index=False)
    print("State names enriched and saved to state_data_enriched.csv")
    return dataframe

# Example usage
df = enrich_state_data(df)
```
